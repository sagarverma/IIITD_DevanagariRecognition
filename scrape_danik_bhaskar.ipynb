```python
htmls = glob.glob('../dataset/danik_bhaskar/*.html')
```

```python
def parse_html(html):
    fin = open(html, 'r')
    data = fin.read()
    fin.close()
    link_found = re.findall(r'digital(.*?)\.PDF', data)
    if link_found:
        pdf_link = "http://digital"+ link_found[0] + ".PDF"
        
        coords_found = re.findall(r'coords="(.*?)" href="java', data)
        coords = []
        for coord in coords_found:
            coords.append(map(float, coord.split(',')))
        
        return [pdf_link, coords]
    return None
```

```python
valid_pdfs = []

for html in htmls:
    parsed = parse_html(html)
    if parsed:
        valid_pdfs.append(parsed + [html])
```
